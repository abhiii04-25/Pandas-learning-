# DataFrames

### Creating a dataframe

```python
import numpy as np 
import pandas as pd
```

```python
data={
    'Name':['Abhishek','Ansh','Meow','Abhi'],
    'Age':[22,22,18,25],
    'City':['Nagpur','Mumbai','Nagpur','Pune'],
    'Salary':[100000,200000,300000,400000]
}
df=pd.DataFrame(data)
```

```python
df
```

```python
data_list=[
    ['Abhishek',22,'Nagpur',100000],
    ['Ansh',22,'Mumbai',200000],
    ['Meow',18,'Nagpur',300000],
    ['Abhi',25,'Pune',400000]
]
columns=["Name","Age","City","Salary"]
df2=pd.DataFrame(data_list,columns=columns)
df2
```

```python
#Selecting and indexing
df2["Name"]
df2[["Name","Age"]]
```

```python
#Creating new colums
df2["Department"]=["IT","HR","Sales","Marketing"]
df2
```

```python
#Removing a column
df2.drop("Department",axis=1,inplace=True)
```

```python
df2
```

```python
#selecting rows
df2.loc[0]
df2.iloc[3]
```

```python
#selecting subset of row and columns
df2.loc[[0,1]][["Name","Age"]]
```

```python
df2.loc[[2,3]][["Name","Salary"]]
```

```python
'''condition selection
I only want those people whose age is greater than 20''' 
df2[df2["Age"]>20]
```

```python
# I only want those people whose age is greater than 20 and from Nagpur
df2[(df2["Age"]>20) & (df2["City"]=="Nagpur")]
```
