```python
#which country has the maximum population
df[df['population'] == df['population'].max()]['country']
```

```python
#what is the capital of the country which has the maximum population
df[df['population'] == df['population'].max()]['capital_city']
```

```python
#which country has the minimum population
df[df['population']==df['population'].min()]['country']
```

```python
#what is the capital of the country which has the minimum population
df[df['population']==df['population'].min()]['capital_city']
```

```python
#Give me top 5 countries with highest democracy index
df.sort_values(by='democracy_score',ascending=False,inplace=True)
```

```python
#how many total regions are there in the dataset
df['region'].value_counts().count()
```

```python
#how many regions lie in the Eastern Europe
df[df['region']=='Eastern Europe']['country']
```

```python
#who is the political leader of the 2nd highest populated country
df[df['population'] == df['population'].nlargest(2).iloc[1]]['political_leader']
```

```python
#how many countries are there whose poplitical leader are unkonwn
df[df['political_leader'].isna()].count()
```

```python
#how many contries have Republic in there name
count=0
def counting(txt):
    global count
    if 'republic' in txt.lower():
        count+=1
    return txt
df['country_long'] = df['country_long'].apply(counting)  
print(count)  
```
