## Merging joining Concadination

```python
import numpy as np
import pandas as pd

```

### Merging 2 Dataframes

```python
employee=pd.DataFrame({
    'employee_id': [1, 2, 3,4,5],
    'Name': ['Alice', 'Bob', 'Charlie','David','Eva'],
    'Age': [25, 30, 35,33,24],
    'Department': ['HR', 'IT', 'Finance','HR','IT'],
})

salaries=pd.DataFrame({
    'employee_id': [1, 2, 3,6,7,8],
    'Salary': [50000, 60000, 70000, 80000,90000,100000],
    'bonus': [5000, 6000, 7000, 8000,9000,10000],
})
```

```python
employee
```

```python
salaries
```

```python
# Merge the two DataFrames on the 'employee_id' column
pd.merge(employee, salaries, on='employee_id', how='inner')
```

```python
# Merge the two DataFrames on the 'employee_id' column using a outer join
pd.merge(employee, salaries, on='employee_id', how='outer')
```

```python
# merge the two DataFrames on the 'employee_id' column using a left join
pd.merge(employee,salaries,on='employee_id', how='left')
```

```python
# Merge the two DataFrames on the 'employee_id' column using a right join
pd.merge(employee,salaries,on='employee_id', how='right')
```

### Concading two DataFrames


```python
df1=pd.DataFrame({
    'A':['A0','A1','A2'],
    'B':['B0','B1','B2'],
    'C':['C0','C1','C2'],
})

df2=pd.DataFrame({
    'A':['A3','A4','A5'],   
    'B':['B3','B4','B5'],
    'C':['C3','C4','C5'],
})
```

```python
df1
```

```python
df2
```

```python
# Concatenate the two DataFrames vertically(colums)
pd.concat([df1,df2])
```

```python
# Concatenate the two DataFrames horizontally(rows)
pd.concat([df1,df2],axis=1)
```

### Joining two DataFrames

```python
df1=pd.DataFrame({
    'Name':['Alice','Bob','Charlie']
},index=[0,1,3])

df2=pd.DataFrame({
    'Age':[25,30,35]
},index=[0,2,4])
```

```python
df1
```

```python
df2
```

```python
#joining two dataframes using join method
df1.join(df2,how='outer')
```

```python
df2.join(df1)
```
