## Missing Data

```python
import numpy as np 
import pandas as pd 
```

### Finding missing values

```python
data={
    'A': [1, 2, np.nan, 4],     
    'B':[5, np.nan, np.nan, 8],
    'C':[9, 10, 11, 12],
    'D':[np.nan, 14, 15, 16]
}
df=pd.DataFrame(data)
```

```python
df
```

```python
# Check for missing values
df.isna()
```

```python
# Count missing values in each column
df.isna().sum()
```

```python
# Check if any missing values exist in each column
df.isna().any()
```

### Removing null values

```python
df
```

```python
# Drop rows with missing values
df.dropna()
```

```python
# Drop rows with less than 3 non-NA values
df.dropna(thresh=3) 
```

### Filling the Missing data

```python
df
```

```python
  # Fill missing values with 0
df.fillna(value=0)
```

```python
# Fill missing values with different values for each column
values={'A':0, 'B':100,'C':200, 'D':400}
df.fillna(value=values)
```

```python
df
```

```python
# Fill missing values with the mean of each column
df.fillna(df.mean())
```
