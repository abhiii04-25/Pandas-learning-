## GroupBy Aggregration

```python
import numpy as np 
import pandas as pd
```

```python
data={
    'category': ['A', 'B', 'A','B', 'A', 'B', 'A', 'B', 'A', 'B'],
    'Stores':['S1','S2','S2','S1','S1','S2','S2','S1','S1','S2'],
    'Sales':[100, 200, 150, 250, 120, 220, 130, 230, 140, 240],
    'Quantity':[10, 20, 15, 25, 12, 22, 13, 23, 14, 24],
    'Date': pd.date_range(start='2023-01-01', periods=10, freq='D')
}

df=pd.DataFrame(data)
df
```

```python
#Group by 'category' and calculate the sum of 'Sales' for each category
df.groupby('category')['Sales'].sum()
```

```python
#Group by stores and claculate the sum of sales for each store
df.groupby('Stores')['Sales'].sum()
```

```python
#Group by multiple columns
df.groupby(['category','Stores'])['Sales'].sum()
```

### Aggregation

```python
df['Sales'].mean()
```

```python
df['Sales'].agg(['sum', 'mean', 'max', 'min','std','count', 'median'])
```
